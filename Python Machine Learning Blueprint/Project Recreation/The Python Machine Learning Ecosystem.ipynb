# The data science/machine learning workflow

- Acquire
- Inspect & Explore
- Clean & Prepare
- Model
- Evaluate
- Deploy

## Acquisition

The data may come from different sources:

- csv file
- pulling down server logs
- building a custom web scraper

## Inspection and exploration

**At this stage, look for the things that are either impossible or highly unlikely.**

- A good practice is to run some simple statistical tests on the data and visualize it.
- It is critical to take note of missing data and address it later during the cleaning and prepartion stage.

## Cleaning and preparation

Steps:

- filter
- aggregate
- impute
- transofrm

This stage highly rely on the tepy of data and the type of library and algorithm utilized.

For example, with natural-language-based text, the transofrmation required will be very different from those required for time series data. 

## Modeling

In this pahse, an appropriate algorithm is slected and a model is trained on the data. 

The basic steps involve splitting the data into training, testing and validation sets. Doing this allows us to get better feedback on how the model will perform in the real word, and prevent from overfitting the model.

## Evaluation

In this stage, we are to understand how well the model makes the predictions.

Types:
- mean-square error
- Euclidean distance
- F1 score

They are alll measure of distance between the actual value and the estimated prediction.

## Deployment

Utlization as a feature within another larger application, a bespoke web application or even just a simple cron job.

```python
import os
import pandas as pd
import requests
from sklearn import datasets

PATH = './datasets/'
iris = pd.read_csv(PATH + 'iris.csv')

iris.head()
iris = iris.rename(columns={'SepalLengthCm': 'sepal length', 'SepalWidthCm': 'sepal width', 'PetalLengthCm': 'petal length', 'PetalWidthCm': 'petal width', 'Species': 'class'})
iris = iris.drop('Id', axis = 1)
```

```python
iris.head()
```

```python
# selecting variable
iris['sepal length']

# data slicing by index
iris.iloc[:3, :2]

# Conditional slcing
iris.loc[:3, [x for x in iris.columns if 'width' in x]]

# List the unique classes for the column
iris['class'].unique()

# Antoher type of conditional slicing
iris[iris['class'] == 'Iris-virginica']

# count the row of dataset
iris.count()
```
